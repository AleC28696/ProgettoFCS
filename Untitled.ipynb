# 1.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
loans_lenders = pd.read_csv('loans_lenders.csv')
loans_lenders.head()
```

```python
loans_lenders_normalized = loans_lenders.assign(lenders=loans_lenders['lenders'].str.split(', ')).explode('lenders')
loans_lenders_normalized
```

# 2.

```python
loans = pd.read_csv('loans.csv')
loans.head()
```

```python
loans[['planned_expiration_time', 'disburse_time']]
```

```python
loans['planned_expiration_datetime'] =  pd.to_datetime(loans['planned_expiration_time'], format='%Y-%m-%d %H:%M:%S.%f %z')
```

```python
loans['disburse_datetime'] =  pd.to_datetime(loans['disburse_time'], format='%Y-%m-%d %H:%M:%S.%f %z')
```

```python
loans['timedelta'] = loans['planned_expiration_datetime'] - loans['disburse_datetime']
```

```python
loans['duration'] = loans['timedelta'].dt.days.astype('Int64')
loans['duration']
```

# 3.

```python
loans_lenders_grouped = loans_lenders_normalized.groupby('lenders', as_index=False).count()
loans_lenders_grouped.columns = ['lenders', 'number_of_fundings']
loans_lenders_grouped = loans_lenders_grouped[loans_lenders_grouped['number_of_fundings'] >= 2]
loans_lenders_grouped
```

# 4.

```python
country_borrowers = loans[['country_name', 'loan_id']].groupby('country_name').count()
country_borrowers.columns = ['number_of_loans']
country_borrowers
```

# 5.

```python
country_loans = loans[['country_name', 'loan_amount']].groupby('country_name', as_index=False).sum()
country_loans
```

# 6.

```python
overall_amount_lent = 0
for i in country_loans['loan_amount']:
    overall_amount_lent += i
```

```python
country_loans['loan_percentage'] = country_loans['loan_amount'] / overall_amount_lent * 100
country_loans
```

# 7.

```python
loans['disburse_year'] = loans['disburse_datetime'].dt.year.astype('Int64')
loans['disburse_year']
```

```python
country_borrowers_per_year = loans[['country_name', 'disburse_year', 'loan_id']].groupby(['country_name', 'disburse_year']).count()
country_borrowers_per_year.columns = ['number_of_loans']
country_borrowers_per_year
```

As point 5

```python
country_loans_per_year = loans[['country_name', 'disburse_year', 'loan_amount']].groupby(['country_name', 'disburse_year'], as_index=False).sum()
country_loans_per_year
```

As point 6

```python
country_loans_per_year['loan_percentage'] = country_loans_per_year['loan_amount'] / overall_amount_lent * 100
country_loans_per_year
```
