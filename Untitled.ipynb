```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
loans_lenders = pd.read_csv('loans_lenders.csv')
loans_lenders.head()
```

# 2.

```python
loans = pd.read_csv('loans.csv')
loans.head()
```

```python
loans[['planned_expiration_time', 'disburse_time']]
```

```python
loans['planned_expiration_datetime'] =  pd.to_datetime(loans['planned_expiration_time'], format='%Y-%m-%d %H:%M:%S.%f %z')
type(loans['planned_expiration_datetime'].loc[0])
```

```python
loans['disburse_datetime'] =  pd.to_datetime(loans['disburse_time'], format='%Y-%m-%d %H:%M:%S.%f %z')
```

```python
loans['timedelta'] = loans['planned_expiration_datetime'] - loans['disburse_datetime']
```

```python
loans['duration'] = loans['timedelta'].dt.days.astype('Int64')
loans.loc[((loans['duration'].notna()) & (loans['duration'] <= 0)), 'duration'] = np.nan  #serve per togliere le differenze negative/nulle 
# Abbiamo deciso di non considerare i loan che hanno duration<=0 nei punti che riguardano gli anni, mentre negli altri sì: il senso è che le date sono state inserite in modo sbagliato ma i loan sono avvenuti
loans['duration'].sort_values(ascending=True)   #abbiamo poi notato che ci sono duration molto basse (tipo un giorno)
```

# 7.

```python
loans['disburse_year'] = loans['disburse_datetime'].dt.year.astype('Int64')
loans['disburse_year']
```

As point 4

```python
disburse_condition = (loans['duration'].notna()) | ((loans['duration'].isna()) & (loans['disburse_time'].notna()))
country_borrowers_per_year = loans[['country_name', 'disburse_year', 'loan_id']][disburse_condition].groupby(['country_name', 'disburse_year']).count()
country_borrowers_per_year.columns = ['number_of_loans']
country_borrowers_per_year
```

As point 5

```python
country_loans_per_year = loans[['country_name', 'disburse_year', 'loan_amount']][disburse_condition].groupby(['country_name', 'disburse_year'], as_index=True).sum()
country_loans_per_year
```

As point 6

```python
percentage_per_country_year = {}

overall_amount_per_year = loans[disburse_condition].groupby('disburse_year')['loan_amount'].sum()

for i in loans[disburse_condition].groupby('country_name')['loan_amount'].sum().index: #i=country, j=year
    percentage_per_country_year[i] = {}
    
for i, j in country_loans_per_year.index:
    percentage_per_country_year[i][j] = {'percentage': (country_loans_per_year.loc[i,j] / overall_amount_per_year.loc[j] * 100)[0]}

percentage_per_country_year
```

```python
df = pd.DataFrame.from_dict({(i,j): percentage_per_country_year[i][j] 
                           for i in percentage_per_country_year.keys() 
                           for j in percentage_per_country_year[i].keys()},
                       orient='index')
```

```python
country_loans_per_year = pd.concat([country_loans_per_year,df], axis=1)
country_loans_per_year.columns = ['loan_amount', 'percentage_per_year']
country_loans_per_year
```

```python
num_lenders = loans_lenders_normalized.groupby('loan_id').count()
num_lenders.columns = ['num_total_lenders']
num_lenders
```

```python
df = loans_lenders_normalized.join(num_lenders, on='loan_id')
df = df.join(loans[['loan_id', 'loan_amount']].set_index('loan_id'), on='loan_id')
df['amount_per_lender'] = df['loan_amount'] / df['num_total_lenders']
money_lent_per_lender = df[['lenders', 'amount_per_lender']].groupby('lenders').sum()
money_lent_per_lender
```

# 9.

```python
lenders = pd.read_csv('lenders.csv')
lenders[lenders['state'].notna()]
```

```python
money_borrowed_per_country = loans[['country_code', 'loan_amount']].groupby('country_code').sum()
money_borrowed_per_country.columns = ['money_borrowed']

df = money_lent_per_lender.join(lenders[['permanent_name', 'country_code']][lenders['country_code'].notna()].set_index('permanent_name'), on='lenders')
df1 = df.groupby('country_code').sum()
df1.columns = ['money_lent']

difference_per_country = money_borrowed_per_country.join(df1, how='outer')
# nel join alcuni stati mancano (nei borrowed o lent), quindi mette NaN e la differenza risulta NaN
difference_per_country = difference_per_country.fillna(0)
difference_per_country['difference'] = difference_per_country['money_lent'] - difference_per_country['money_borrowed']
difference_per_country
```

# 10.

```python
country_stats = pd.read_csv("country_stats.csv")

country_ratio = difference_per_country.join(country_stats[['country_code', 'population']].set_index('country_code'), on='country_code', how='inner')
country_ratio['difference/population'] = country_ratio['difference'] / country_ratio['population']
country_ratio.sort_values('difference/population', ascending=False)
```

# 11.

```python
country_ratio2 = country_ratio.join(country_stats[['country_code', 'population_below_poverty_line']].dropna().set_index('country_code'), on='country_code', how='inner')
country_ratio2['difference/wealthy_population'] = country_ratio2['difference'] / (country_ratio2['population'] * ((100-country_ratio2['population_below_poverty_line']))/100)
country_ratio2.sort_values('difference/wealthy_population', ascending=False)
```

```python
loans.columns
```

```python
loans['planned_expiration_year'] = loans['planned_expiration_datetime'].dt.year.astype('Int64')
loans[['planned_expiration_year']]
```

```python
loans1 = loans[['loan_amount', 'planned_expiration_datetime', 'planned_expiration_year', 'disburse_datetime', 'disburse_year', 'duration']][loans['duration'].notna()]
loans1
```

```python
max_year = loans1['planned_expiration_year'].max()
min_year = loans1['disburse_year'].min()
```

```python
year_amount = {}
for i in range(min_year, max_year+1):
    year_amount[i]=0
year_amount
```

```python
for index in loans1.index:
    
    if(loans1['planned_expiration_year'].loc[index] == loans1['disburse_year'].loc[index]):
        year_amount[loans1['disburse_year'].loc[index]] += loans1['loan_amount'].loc[index]
        
    else:
        
        for year in range(loans1['disburse_year'].loc[index], loans1['planned_expiration_year'].loc[index]+1):
            
            if(year == loans1['disburse_year'].loc[index]):
                
                last_day = datetime.date(year=loans1['disburse_year'].loc[index], month=12, day=31)
                days = (last_day - loans1['disburse_datetime'].loc[index].date()).days + 1
                
            elif(year == loans1['planned_expiration_year'].loc[index]):
                
                first_day = datetime.date(year=loans1['planned_expiration_year'].loc[index], month=1, day=1)
                days = (loans1['planned_expiration_datetime'].loc[index].date() - first_day).days + 1
                
            else:
                
                first_day = datetime.date(year=loans1['planned_expiration_year'].loc[index], month=1, day=1)
                last_day = datetime.date(year=loans1['disburse_year'].loc[index], month=12, day=31)
                days = (last_day - first_day).days + 1
            
            year_amount[year] += loans1['loan_amount'].loc[index] * days / loans1['duration'].loc[index]
```

```python
year_amount = pd.Series(year_amount)
year_amount
```
