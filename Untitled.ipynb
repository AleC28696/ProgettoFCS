```python
import pandas as pd
import numpy as np
import datetime
```

Importazione del dataset "loans_lenders.csv":

```python
loans_lenders = pd.read_csv('loans_lenders.csv')
loans_lenders.head()
```

Attraverso la funzione "str.split", le stringhe della colonna "lenders" vengono splittate ad ogni occorrenza della sequenza di caratteri ", ", ottenendo per ogni cella un vettore. # 2.

```python
loans = pd.read_csv('loans.csv')
loans.head()
```

```python
loans[['planned_expiration_time', 'disburse_time']]
```

Conversione delle due colonne in formato datetime:

```python
loans['planned_expiration_datetime'] = pd.to_datetime(loans['planned_expiration_time'], format = '%Y-%m-%d %H:%M:%S.%f %z')
```

```python
loans['disburse_datetime'] = pd.to_datetime(loans['disburse_time'], format = '%Y-%m-%d %H:%M:%S.%f %z')
```

Si calcola la differenza tra le due date (la differenza è NA se almeno una delle due date è missing) e successivamente si estrae da tale differenza solo il numero di giorni, convertendola in "int".

```python
loans['timedelta'] = loans['planned_expiration_datetime'] - loans['disburse_datetime']
loans['duration'] = loans['timedelta'].dt.days.astype('Int64') # estrae i giorni
loans['duration']
```

Come si può notare dal risultato sottostante, sono presenti valori negativi o nulli di 'duration':

```python
loans['duration'][((loans['duration'].notna()) & (loans['duration'] <= 0))]
```

Si procede quindi alla correzione delle 'duration' negative o nulle, sostituendo tali valori con NaN:

```python
loans.loc[((loans['duration'].notna()) & (loans['duration'] <= 0)), 'duration'] = np.nan
```

Come si può notare, ci sono duration molto basse, ad esempio di un solo giorno, probabilmente dovute a errori nella compilazione del dataset:

```python
loans['duration'].sort_values(ascending = True)
```

Abbiamo deciso di non considerare i loan che hanno duration<=0 nei successivi punti del progetto che riguardano gli anni, impostando la 'duration' a NA, mentre negli altri punti sì: questo perché le date sono state inserite in modo sbagliato, ma i loan sono comunque avvenuti.

### 7.4

Si crea una condizione logica "disburse_condition": essa è True solo se il valore di 'duration' nella tabella 'loans' è diversa da NA (ossia quando 'disburse_time' e 'planned_expiration_time' sono entrambe non NA) oppure quando 'disburse_time' è non NA. Questo serve per eliminare i loan con date inserite in modo non corretto, come già anticipato precedentemente, e i loan con 'disburse_time' mancante, in quanto per questi ultimi non è possibile conoscere il disburse year.

```python
disburse_condition = (loans['duration'].notna()) | ((loans['duration'].isna()) & (loans['disburse_time'].notna()))
```

Si selezionano da loans le righe che rispettano la condizione "disburse_condition" e si effettua un raggruppamento a due livelli: prima in base a 'country_name', poi rispetto al 'disburse_year'. Infine, si contano le righe in ogni gruppo.

```python
country_borrowers_per_year = loans[['country_name', 'disburse_year', 'loan_id']][disburse_condition].groupby(['country_name', 'disburse_year']).count()
country_borrowers_per_year.columns = ['number_of_loans']
country_borrowers_per_year
```

### 7.5

Anche in questo caso si selezionano le righe attraverso la condizione "disburse_condition". Si effettua poi un raggruppamento a due livelli identico al precedente. Infine, si sommano i "loan_amount" in ogni gruppo.

```python
country_loans_per_year = loans[['country_name', 'disburse_year', 'loan_amount']][disburse_condition].groupby(['country_name', 'disburse_year']).sum()
country_loans_per_year
```

### 7.6

Si crea innanzitutto un dizionario che ha come chiavi i nomi dei Paesi e come valore per ogni chiave ha un sotto-dizionario.
Ogni sotto-dizionario ha come chiavi gli anni e per ogni anno ha come valore un altro sotto-sotto-dizionario.
Ognuno di questi sotto-sotto-dizionari ha una sola chiave 'percentage', a cui corrisponde come valore la percentuale di loan_amount per quel Paese in quello specifico anno, calcolata relativamente all'overall amount di tutti i Paesi in quell'anno.

```python
percentage_per_country_year = {}

# Crea una serie Pandas con gli importi totali ricevuti ogni anno da tutti i Paesi 
overall_amount_per_year = loans[disburse_condition].groupby('disburse_year')['loan_amount'].sum()

# i=country, j=year

# Crea un sotto-dizionario (vuoto) per ogni Paese:
for i in loans[disburse_condition].groupby('country_name')['loan_amount'].sum().index:
    percentage_per_country_year[i] = {}

# All'interno dei sotto-dizionari appena creati, crea altri sotto-sotto-dizionari per ogni anno:
for i, j in country_loans_per_year.index:
    percentage_per_country_year[i][j] = {'percentage': (country_loans_per_year.loc[i,j] / overall_amount_per_year.loc[j] * 100)[0]}

# Si stampa il valore corrispondente alla prima chiave del dizionario (Afghanistan), a titolo di esempio:
percentage_per_country_year['Afghanistan']
```

A questo punto, partendo dal dizionario "percentage_per_country_year" si crea un DataFrame Pandas:

```python
df = pd.DataFrame.from_dict({(i,j): percentage_per_country_year[i][j] 
                           for i in percentage_per_country_year.keys() 
                           for j in percentage_per_country_year[i].keys()},
                       orient='index')
df
```

Infine, si concatena il dataframe appena creato con "country_loans_per_year":

```python
country_loans_per_year = pd.concat([country_loans_per_year, df], axis = 1)
country_loans_per_year
```

```python
lenders = pd.read_csv('lenders.csv')
lenders
```

Si effettua un group by sulla tabella 'loans' in base a 'country_code', con una somma sui 'loan_amount'. Questo permette di calcolare il totale di soldi ricevuti in prestito da ogni Paese.

```python
money_borrowed_per_country = loans[['country_code', 'loan_amount']].groupby('country_code').sum()
money_borrowed_per_country.columns = ['money_borrowed']
money_borrowed_per_country
```

Tramite un join, si aggiunge il codice del Paese alla tabella 'money_lent_per_lender'. Successivamente, si effettua un group by in base a tale codice, con una somma degli importi in ogni gruppo, in modo da ottenere l'importo totale prestato da ogni Paese.

```python
df = money_lent_per_lender.join(lenders[['permanent_name', 'country_code']][lenders['country_code'].notna()].set_index('permanent_name'), on='lenders')
df = df.groupby('country_code').sum()
df.columns = ['money_lent']
df
```

Si crea quindi una nuova tabella 'difference_per_country', risultato di un join tra 'money_borrowed_per_country' e 'df'. Tuttavia, non tutti i Paesi hanno sia ricevuto che effettuato prestiti, quindi nelle due tabelle non sono presenti gli stessi Stati. Si effettua quindi un outer join, che unisce le due tabelle assegnando il valore NaN nel caso in cui i valori 'money_borrowed' o 'money_lent' non siano noti per quello Stato. Successivamente, si sostituisce a ogni NaN il valore 0, intendendo che il Paese non ha prestato/ricevuto nessun importo.

```python
difference_per_country = money_borrowed_per_country.join(df, how='outer')
difference_per_country = difference_per_country.fillna(0)
```

Si calcola infine la differenza tra 'money_borrowed' e 'money_lent':

```python
difference_per_country['difference'] = difference_per_country['money_lent'] - difference_per_country['money_borrowed']
difference_per_country
```

Tramite un join tra le tabelle 'difference_per_country' e 'country_stats', si aggiunge alla prima la popolazione di ogni Paese. # 12.

Si estrae l'anno dalla colonna 'planned_expiration_datetime':

```python
loans['planned_expiration_year'] = loans['planned_expiration_datetime'].dt.year.astype('Int64')
loans[['planned_expiration_year']]
```

Si selezionano solo alcune colonne della tabella 'loans' e si considerano soltanto i loan con date inserite correttamente: questo viene effettuato imponendo sulle righe la condizione logica loans['duration'].notna().

```python
loans1 = loans[['loan_amount', 'planned_expiration_datetime', 'planned_expiration_year', 'disburse_datetime', 'disburse_year', 'duration']][loans['duration'].notna()]
loans1
```

Si identificano l'anno più recente nella colonna 'planned_expiration_year' e quello meno recente nella colonna 'disburse_year':

```python
max_year = loans1['planned_expiration_year'].max()
min_year = loans1['disburse_year'].min()
```

Si crea un dizionario che possiede come chiavi tutti gli anni compresi tra 'min_year' e 'max_year'. I valori vengono tutti impostati a 0: successivamente verranno aggiornati in modo che, alla fine, ogni valore corrisponda al total amount of loans per lo specifico anno.

```python
year_amount = {}
for i in range(min_year, max_year+1):
    year_amount[i] = 0
year_amount
```

Per tenere conto della presenza di anni bisestili, si crea un dizionario contenente il numero di giorni di ogni anno considerato:

```python
days_per_year = year_amount

for i in days_per_year.keys():
    first_day = datetime.date(year=i, month=1, day=1)
    last_day = datetime.date(year=i, month=12, day=31)
    days_per_year[i] = (last_day - first_day).days + 1
    
days_per_year
```

Attraverso un ciclo for, si calcola quindi il total amount of loans per ogni anno:

```python
# 'index' corre sui valori della chiave della tabella 'loans1', ossia i numeri di riga della tabella: esso identifica
# quindi i vari loan

for index in loans1.index:
    
    # Si verifica innanzitutto se il 'disburse_year' e il 'planned_expiration_year' dell'attuale loan coincidono: se ciò
    # è vero, significa che il loan è avvenuto all'interno di un solo anno, pertanto è sufficiente aggiungere l'intero
    # 'loan_amount' al valore della corrispondente chiave nel dizionario 'year_amount'.
    
    if(loans1['planned_expiration_year'].loc[index] == loans1['disburse_year'].loc[index]):
        
        year_amount[loans1['disburse_year'].loc[index]] += loans1['loan_amount'].loc[index]
        
        
    # Se invece i due anni sono diversi, si procede con la suddivisione del 'loan_amount' negli anni interessati dal loan:
    
    else: # TRASFORMARLO IN UN elif(disburse_year < planned_expiration_year) !!!!!!!!!!!!!!!!!!!!!!
        
        # Si entra in un sotto-ciclo for, in cui 'year' corre sugli anni interessati dal loan. Per ogni anno si calcolano
        # i giorni interessati dal loan (tramite un if-elif-else) e successivamente la frazione di 'loan_amount' da
        # attribuire a quell'anno. Si aggiunge quindi questa quantità al dizionario 'year_amount', aggiornando il valore
        # per quell'anno.
        for year in range(loans1['disburse_year'].loc[index], loans1['planned_expiration_year'].loc[index] + 1):
            
            # Se 'year' corrisponde al primo anno del loan (ovver 'disburse_year'), si calcola il numero di giorni come la
            # differenza tra il primo giorno del loan e il 31 dicembre di quell'anno
            if(year == loans1['disburse_year'].loc[index]):
                
                # 'last_day' è il 31 dicembre dell'anno 'year'
                last_day = datetime.date(year = loans1['disburse_year'].loc[index], month = 12, day = 31)
                days = (last_day - loans1['disburse_datetime'].loc[index].date()).days + 1
                
            # Se 'year' corrisponde all'ultimo anno del loan (ovver 'planned_expiration_year'), si calcola il numero di
            # giorni come la differenza tra l'1 gennaio di quell'anno e l'ultimo giorno del loan
            elif(year == loans1['planned_expiration_year'].loc[index]):
                
                # 'first_day' è il 1 gennaio dell'anno 'year'
                first_day = datetime.date(year=loans1['planned_expiration_year'].loc[index], month=1, day=1)
                days = (loans1['planned_expiration_datetime'].loc[index].date() - first_day).days + 1
                
            # Se l'anno 'year' è un anno intermedio tra il primo e l'ultimo anno del loan, significa che l'intero anno
            # 'year' è interessato dal loan in questione. Dunque è sufficiente utilizzare il numero di giorni dell'anno,
            # calcolati precedentemente
            else:
                
                days = days_per_year[year]
            
            year_amount[year] += loans1['loan_amount'].loc[index] * days / loans1['duration'].loc[index]
```

Infine, si converte il dizionario in una serie Pandas:

```python
year_amount = pd.Series(year_amount)
year_amount
```
